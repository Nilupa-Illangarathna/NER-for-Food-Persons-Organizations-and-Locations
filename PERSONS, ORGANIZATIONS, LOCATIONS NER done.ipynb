```python
!pip install transformers

from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("Babelscape/wikineural-multilingual-ner")
model = AutoModelForTokenClassification.from_pretrained("Babelscape/wikineural-multilingual-ner")

nlp = pipeline("ner", model=model, tokenizer=tokenizer, grouped_entities=True)



example = """In the culinary world, there's an exciting collaboration between two renowned organizations, Gourmet Delights and Sweet Treats Bakery (STB). Gourmet Delights, known for its delectable range of premium food products, has teamed up with STB, a popular bakery chain with a reputation for crafting mouthwatering desserts.

The partnership aims to create a fusion of flavors that will delight food enthusiasts worldwide. Gourmet Delights will contribute its expertise in sourcing high-quality ingredients from various regions, including India, to infuse exotic flavors into STB's delectable pastries and cakes.

One of the highlights of this collaboration is the introduction of the "Kandy Delight" - a new line of cupcakes inspired by the flavors of the beautiful city of Kandy in Sri Lanka. These cupcakes will feature a unique blend of aromatic spices and tropical fruits sourced directly from the region.

Moreover, the talented pastry chef, John, renowned for his innovative recipes, will be joining the creative team. John's flair for experimenting with flavors will undoubtedly add a distinct touch to the new line of desserts. His expertise in creating delectable combinations has earned him a stellar reputation in the culinary world.

To further expand their offerings, Gourmet Delights and STB are exploring partnerships with local artisans in India to incorporate traditional ingredients into their products. This initiative will not only support local farmers but also introduce customers to the authentic tastes of India.

The collaboration between Gourmet Delights, Sweet Treats Bakery, India, Kandy, and John promises to be a delightful journey for food enthusiasts. With a blend of international flavors and the expertise of seasoned chefs, this partnership is set to redefine the culinary landscape and satiate the cravings of food lovers everywhere."""

ner_results = nlp(example)
for i in range(len(ner_results)):
  print(ner_results[i])
```
